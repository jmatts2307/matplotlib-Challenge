```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
#In order to get the results looking like the preview it came with, 'study_results' should come first when merging, 
#followed by 'mouse_metadata', merging on 'Mouse ID'.
complete_mouse_data = pd.merge(study_results, mouse_metadata, on="Mouse ID")

# Display the data table for preview
complete_mouse_data.head()
```

```python
# Checking the number of mice.
#Use 'nunique()' to get the number of unique mice, it is an easier approach to
#directly getting the number of mice in the "Mouse ID" column
number_of_mice = complete_mouse_data["Mouse ID"].nunique()
number_of_mice
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint.
#using the 'duplicated()' on the dataset helps to check for duplicate rows on 'Mouse ID' and 'Timepoint',
#and using '.loc' extracts the 'Mouse ID' of the duplicated rows
duplicate_mice = complete_mouse_data.loc[complete_mouse_data.duplicated(subset=["Mouse ID", "Timepoint"]),"Mouse ID"].unique()
duplicate_mice
```

```python
# Optional: Get all the data for the duplicate mouse ID.
#Once again using '.loc', applied it to extract all rows where 'Mouse ID' is in 'duplicate_mice',
#And using 'isin()' checks if the mouse is present in 'duplicate_mice', and returns matches
duplicate_mouse_data = complete_mouse_data.loc[complete_mouse_data["Mouse ID"].isin(duplicate_mice)]
duplicate_mouse_data
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
#Used '.isin' function in 'duplicate_mice' to check if 'Mouse ID' is in its list, then used the index
#function to retrieve those rows, and used the 'drop' function to drop those rows from the dataframe
duplicate_rows = complete_mouse_data[complete_mouse_data["Mouse ID"].isin(duplicate_mice)].index
clean_mouse_data = complete_mouse_data.drop(duplicate_rows)
clean_mouse_data.head()
```

```python
# Checking the number of mice in the clean DataFrame.
cleaned_mice_number = clean_mouse_data["Mouse ID"].nunique()
cleaned_mice_number
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen:
# mean, median, variance, standard deviation, and SEM of the tumor volume.
# Assemble the resulting series into a single summary DataFrame.
drug_regimen = clean_mouse_data.groupby("Drug Regimen")["Tumor Volume (mm3)"]

mean = drug_regimen.mean()
median = drug_regimen.median()
var = drug_regimen.var()
std = drug_regimen.std()
sem = drug_regimen.sem()

summary_statistics_df = pd.DataFrame({
    "Mean Tumor Volume": mean,
    "Median Tumor Volume": median, 
    "Tumor Volume Variance": var,
    "Tumor Volume Std. Dev.": std,
    "Tumor Volume Std. Err.": sem
})

summary_statistics_df
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)
#had to look at this for reference on how to use '.agg' method: https://stackoverflow.com/questions/53781634/aggregation-in-pandas
# Using the aggregation method, produce the same summary statistics in a single line
summary_statistic_agg = clean_mouse_data.groupby(["Drug Regimen"])[["Tumor Volume (mm3)"]].agg(['mean', 'median', 'var', 'std', 'sem'])
summary_statistic_agg
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:
# Capomulin, Ramicane, Infubinol, and Ceftamin
#Grouped the rows to only include the ones that are sepcified in the 4 treatment regimens
treatment_regimens = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]
filtered_data = clean_mouse_data[clean_mouse_data["Drug Regimen"].isin(treatment_regimens)]
# Start by getting the last (greatest) timepoint for each mouse
#Grouped the filtered data by ID, and got the maximum timepoint for each mouse which is the last 
#recorded observation, and the resetindex gives the result as dataframe and not an index
last_timepoint = filtered_data.groupby("Mouse ID")["Timepoint"].max().reset_index()

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
merged_data = pd.merge(last_timepoint, filtered_data, on=["Mouse ID", "Timepoint"])
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
tumor_vol_data = []
#looping through each treatment regimen, and locating rows that contain mice on each drug and getting tumor volumes
for regimen in treatments:

    # Locate the rows which contain mice on each drug and get the tumor volumes
    regimen_data = merged_data[merged_data["Drug Regimen"] == regimen]["Tumor Volume (mm3)"]

    # add subset
    tumor_vol_data.append(regimen_data)
# Calculate the IQR and quantitatively determine if there are any potential outliers.
#Referenced back to an example we did in class to figure out quartiles Module 5.3 activity 2
    quartiles = regimen_data.quantile([0.25, 0.5, 0.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq - lowerq
    
    # Determine outliers using upper and lower bounds
    lower_bound = lowerq - (1.5 * iqr)
    upper_bound = upperq + (1.5 * iqr)
    outliers = regimen_data[(regimen_data < lower_bound) | (regimen_data > upper_bound)]

    print(f"{regimen}'s potential outliers: {outliers}")
```

```python
# Generate a box plot that shows the distribution of the tumor volume for each treatment group.
plt.boxplot(tumor_vol_data, labels=treatments, patch_artist=False)

plt.title("Distribution of Final Tumor Volume for Each Treatment Regimen")
plt.xlabel("Drug Regimen")
plt.ylabel("Final Tumor Volume (mm3)")

plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model
# for mouse weight and average observed tumor volume for the entire Capomulin regimen
#Used 'st.pearson' to calculate the Pearson correlation coefficient and p-value
#Used https://youtu.be/lapKRrSdn_U for reference on getting the linear regression line
capomulin_data = clean_mouse_data[clean_mouse_data["Drug Regimen"] == "Capomulin"]
average_tumor_volume = capomulin_data.groupby("Mouse ID").mean(numeric_only=True)

x_values = average_tumor_volume["Weight (g)"]
y_values = average_tumor_volume["Tumor Volume (mm3)"]

correlation = st.pearsonr(x_values, y_values)
print(f"The correlation coefficient between mouse weight and average tumor volume is {correlation[0]}")
(slope, intercept, rvalue, pvalue, stderr) = st.linregress(x_values, y_values)

regress_values = slope * x_values + intercept

plt.scatter(x_values, y_values, color="blue")
plt.plot(x_values, regress_values, color="red")

plt.title("Mouse Weight Vs. Average Observed Tumor Volume For Capomulin")
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")

plt.show()
```
